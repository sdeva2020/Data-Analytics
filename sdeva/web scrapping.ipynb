```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.imdb.com/list/ls033255925/"
```

```python
a=requests.get(url)
```

```python
a
```

```python
b=BeautifulSoup(a.text)
```

```python
b
```

```python
moviesorder=b.select(".lister-item-header a")
```

```python
moviesorder
```

```python
year=b.select(".unbold")
```

```python
year
```

```python
import pandas as pd
```

```python
ratings=b.select(".ipl-rating-star.small .ipl-rating-star__rating")
```

```python
ratings
```

```python
barbiemovielist=pd.DataFrame(moviesorder,columns=["List"])
```

```python
barbiemovielist
```

```python
barbiemovielist["Ratings"]=ratings
```

```python
barbiemovielist
```

```python
votes=b.select(".text-muted .text-muted+ span")
```

```python
votes
```

```python
barbiemovielist["votes"]=votes
```

```python
barbiemovielist
```

```python
director=b.select(".text-muted a:nth-child(1)")
```

```python
director
```

```python
barbiemovielist["Directors"]=director
```

```python
barbiemovielist
```

```python
barbiemovielist.to_csv("D:/vac/Barbiemovielist.csv")
```
